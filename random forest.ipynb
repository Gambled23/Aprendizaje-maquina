```python
#Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns

from sklearn import tree

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
df = pd.read_csv('datasets/pcos_Modified.csv', sep=";")
#df.info()
#df.columns
#print percentage of missing values for each column
res = df.columns
for col in res:
    print(col, df[col].isna().mean())
```

```python
sns.pairplot(df, hue="PCOS_Diagnosis")
print(f"the size of the dataset before cleaning the data is {df.shape}")
df = df.dropna(how='any',axis=0)
print(f"the size of the dataset after cleaning the data is {df.shape}")
```

#### 2 opciones para seleccionar variables

```python
# variablex = df.iloc[:,0:5].values #todos los renglones de la columna 0 a la columna 5, excluyendo la columna 5
# variabley = df.iloc[:,5] #todos los renglones de la columna 5

variablex = df[["Age","BMI", "Menstrual_Irregularity", "Testosterone_Level(ng/dL)", "Antral_Follicle_Count"]].values
variabley = df["PCOS_Diagnosis"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(variablex, variabley, test_size=0.2, random_state=42)
print(f"El tamaño del set de entrenamiento \nx: {x_train.shape} \ny: {y_train.shape}")
print(f"El tamaño del set de prueba \nx: {x_test.shape} \ny: {y_test.shape}")
```

```python
# model = RandomForestClassifier(random_state=42, oob_score=True)
model = RandomForestClassifier(criterion='gini',
                               max_features='sqrt',
                               bootstrap=True,
                               max_samples=2/3,
                               oob_score=True, 
                               max_depth=5)
print(model.get_params())
model.fit(x_train, y_train)
```

```python
print(len(model))
print(f"El puntaje obtenido es: {model.oob_score_*100}%")

#Checar los arboles generados por el modelo
# for arbol in model.estimators_:
#   tree.plot_tree(arbol, feature_names = df.columns[:-1])
#   plt.show()
```

```python
y_pred = model.predict(x_test)
print(y_pred[0:10])
f1 = f1_score(y_test, y_pred, average='macro')
print(f"El puntaje F1 obtenido es: {f1}")
percent = model.score(x_test, y_test)
print(f"El porcentaje de aciertos es: {percent*100}%")
```
